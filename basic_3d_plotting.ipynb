```python
import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D

# Create a 3d axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Center a local frame at 2,2,2
center = [2, 2, 2]
ax.plot([center[0],3], [center[1],2], [center[2],2] )
ax.plot([center[0],2], [center[1],3], [center[2],2] )
ax.plot([center[0],2], [center[1],2], [center[2],3] )

# Fix axis limits
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])
plt.show()
```
